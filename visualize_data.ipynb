```python
import numpy as np
import matplotlib.pyplot as plt
import torch
import torchvision
```

```python
# This is model 6 Constrained

data = np.load('/Users/zhangfengzhe/Desktop/urop_generative_model/Constrainted-NVAE/expr/eval-15/gpu_0_samples_20.npz')
for key in data.keys():
    data_point = data[key]
images = data_point[0:32, :, :, :]

def imshow(img):
    npimg = img.numpy()
    plt.imshow(np.transpose(npimg, (1, 2, 0)))
    plt.show()

# show images
imshow(torchvision.utils.make_grid(torch.tensor(images)))
```
